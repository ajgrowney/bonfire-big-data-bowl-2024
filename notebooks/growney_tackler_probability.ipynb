```python
plays_df = pd.read_csv(f"{DATA_ROOT}/plays.csv")
player_pos = pd.read_csv(f"{DATA_ROOT}/players.csv")
tackles_df = pd.read_csv(f"{DATA_ROOT}/tackles.csv")
```

```python
tackles_with_pos_df = pd.merge(tackles_df, player_pos, on="nflId", how="left")
plays_with_tackles_df = pd.merge(plays_df, tackles_with_pos_df, on=["playId", "gameId"], how="left")
```
